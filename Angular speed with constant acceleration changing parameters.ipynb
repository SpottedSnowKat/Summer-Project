```python
%matplotlib inline

import numpy as np 
import matplotlib.pyplot as plt
```

```python
def speed(t,a,r):
    ratio = a/r
    w = ratio*t/(1+(ratio**2)*(t**4)*0.25)
    return w

for i in range(-20, 20):
    print(speed(i,20,10))
```

```python
list1 = [speed(i,20,10) for i in range(-20,20)]
list2 = [speed(i,20,20) for i in range(-20,20)]
list3 = [speed(i,20,30) for i in range(-20,20)]
list4 = [speed(i,10,20) for i in range(-20,20)]
list5 = [speed(i,20,20) for i in range(-20,20)]
list6 = [speed(i,30,20) for i in range(-20,20)]
```
